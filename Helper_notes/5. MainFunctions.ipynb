**Copy this into your `MainWindow` class to handle the data flow:**

```python
def handle_search(self):
    # Get the text the user typed
    item_name = self.search_input.text()

    # 1. Ask the Heart (Fetch historical data)
    df = database.get_price_history(item_name)

    # 2. Tell the Legs (Update the Map)
    # This sends your Python data across the bridge to the JavaScript map!
    self.map_widget.update_map(df)

    # 3. Tell the Hands (Draw the Chart)
    self.chart_widget.update_chart(df)

    # 4. Ask the Brain (ML Prediction)
    prediction = ml_model.predict_price(item_name, "Rewe", 0)
    print(f"üß† Intelligence Report: Predicted future price is ‚Ç¨{prediction:.2f}")
```
